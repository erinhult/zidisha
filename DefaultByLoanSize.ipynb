```python
import pymysql as mdb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
import datetime 

```

```python
db = mdb.connect(user="root", host="localhost", db="zidisha", charset='utf8')
dfloans = pd.read_sql("SELECT * from loanapplic  ", db)

dfbids = pd.read_sql("SELECT * from loanbids  ", db)

dflenders = pd.read_sql("SELECT * from lenders  ", db)
#Alternately, could run more specific SQL queries rather than using 
#   pandas to select from & join tables. There is an example of where I used SQL querying
#   further south in this notebook.
```

Issue: Loan requests from returning borrowers is exceeding lending levels
    - Based on lender retention data, there are opportunities to promote increased lending/engagement, but this is harder to influence than adjusting he # of loan applications
    - Zidisha could meet lending level with only returning borrowers
    - From Jack/Bayes Impact, returning borrowers tend to have lower credit risk than new
    - Thus if credit risk was only factor in selection, there would be no/few new borrowers
Objectives:
    - Provide Zidisha information regarding default rate by loan size to adjust loan size guidelines for returning borrowers
    - Highlight that default rate amongst returning borrowers is significant and that tighter criteria for returning borrowers could potentially reduce losses (connects to Jack's project)



```python
ddates=dfloans[['borrowerid','loanid','expires','applydate','AcceptDate','ActiveDate','RepaidDate','auth_date','active','reqdamt']]
ddates.columns=['borrowerid','loanid','expires','applydate','AcceptDate','ActiveDate','RepaidDate','auth_date','loanstat','reqdamt']

ddates['aptoac']=(ddates['AcceptDate'].apply(float)-ddates['applydate'].apply(float))/86400
ddates['aptoex']=(ddates['expires'].apply(float)-ddates['applydate'].apply(float))/86400
##ddates['acORex']=[max([ddates['expires'][x],ddates['AcceptDate'][x]]) for x in ddates.index]
ddates['acORex']=[ddates['expires'][x] if ddates['loanstat'][x]==7 else\
                  ddates['AcceptDate'][x] for x in ddates.index]

#add loan status to bid database
dfloansS=dfloans[['loanid','active','borrowerid']]
dfloansS.columns = ['loanid','loanstat','borrowerid']
dfbids=pd.merge(dfbids, dfloansS, left_on='loanid',right_on='loanid',how='left')
```

Build dataframe of quantities to use as features or key reporting metrics.
Note: Time frame can be adjusted.

```python
startT=(datetime.datetime(2010,1,1))
NW=365/(12*7)  # weeks per year
NY=4.9 # number of years to include in dataframe
monthlyDat=pd.DataFrame([time.mktime((startT+datetime.timedelta(weeks=x*NW)).timetuple()) \
                      for x in range(int((365/7)*NY/NW))])
# Two time periods are listed: period 1 and then period 2 is the following period 
# (typically period is a month, but you could run analyis on a weekly time scale if desired)
# Two months are included 
monthlyDat.columns=['per1End'] 
monthlyDat['per1Start']=monthlyDat['per1End']-NW*7*86400
monthlyDat['per2End']=monthlyDat['per1End']+NW*7*86400
monthlyDat['P1Sdate']=pd.DataFrame([startT+datetime.timedelta(weeks=x*NW) \
                      for x in range(int((365/7)*NY/NW))])
```

```python
# Some loan application records have problematic dates for acceptance or expiration: 
#   -> pull only records with non-zero, finite dates
ddates=ddates[ddates['acORex']>10]
```

```python
# 'numOpen' calculates the number of open loans at a the end of period 1 ('per1End')
monthlyDat['numOpen']=[sum((ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['acORex']>monthlyDat['per1End'][x])) for x in monthlyDat.index]
```

```python
#What percentage of funding to expired loans receive?
ddates['bidtotal']=[sum(dfbids['bidamount'][dfbids['loanid']==ddates['loanid'][x]])\
                                       for x in ddates.index]

monthlyDat['fundFractionExpire1']=[mean((ddates['bidtotal']/ddates['reqdamt'])\
                                        [(ddates['expires']<monthlyDat['per1End'][x])&
                         (ddates['expires']>monthlyDat['per1Start'][x])&\
                         (ddates['loanstat']==7)])*1.0 for x in monthlyDat.index]
```

```python
# How many $ of accepted loans in period 1 (prev month)?
monthlyDat['amtAccept1']=[sum(ddates['reqdamt'][(ddates['AcceptDate']<monthlyDat['per1End'][x])&
                         (ddates['AcceptDate']>monthlyDat['per1Start'][x])&\
                         (ddates['loanstat']!=7)]) for x in monthlyDat.index]
# NB: variables appended '2' are just a shifted version of 1. If storage was a concern, would not store this in the 
#     database and would just use offset rows of 1, but just including as a separate column for simplicity
monthlyDat['amtAccept2']=[sum(ddates['reqdamt'][(ddates['AcceptDate']<monthlyDat['per2End'][x])&
                         (ddates['AcceptDate']>monthlyDat['per1End'][x])&\
                         (ddates['loanstat']!=7)]) for x in monthlyDat.index]
# What $ amount of loans expired (bid or unbid)
monthlyDat['amtExpired1']=[sum(ddates['reqdamt'][(ddates['expires']<monthlyDat['per1End'][x])&
                         (ddates['expires']>monthlyDat['per1Start'][x])&\
                         (ddates['loanstat']==7)]) for x in monthlyDat.index]
monthlyDat['amtExpired2']=[sum(ddates['reqdamt'][(ddates['expires']<monthlyDat['per2End'][x])&
                         (ddates['expires']>monthlyDat['per1End'][x])&\
                         (ddates['loanstat']==7)]) for x in monthlyDat.index]
#What $ amount of loans was requested in period 1
monthlyDat['amtRequest1']=[sum(ddates['reqdamt'][(ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x])]) for x in monthlyDat.index]
#What $ amount of loans was requested in period 2
monthlyDat['amtRequest2']=[sum(ddates['reqdamt'][(ddates['applydate']<monthlyDat['per2End'][x])&
                         (ddates['applydate']>monthlyDat['per1End'][x])]) for x in monthlyDat.index]
#Expiration rate for period 2 ($ exp / ($exp+$accept))
monthlyDat['expRate2']=monthlyDat['amtExpired2']/(monthlyDat['amtExpired2']+monthlyDat['amtAccept2'])

#How many $ were bid during period 1?
dfbids['biddate']=dfbids['biddate'].apply(float)
monthlyDat['amtBid1']=[sum(dfbids['bidamount'][(dfbids['biddate']<monthlyDat['per1End'][x])&\
                         (dfbids['biddate']>monthlyDat['per1Start'][x])]) \
                          for x in monthlyDat.index]
#Was this the first loan?
ddates['firstLoan']=[min(ddates['applydate'][ddates['borrowerid'][x]==ddates['borrowerid']])==\
                     ddates['applydate'][x] for x in ddates.index]
#What fraction of $ requested was from new borrowers in Period 1?
monthlyDat['fracNew1']=[sum(ddates['reqdamt'][(ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x])&\
                         (ddates['firstLoan']==True)])/(monthlyDat['amtRequest1'][x]) \
                         for x in monthlyDat.index]
#What fraction of borrowers were new in Period 1?
monthlyDat['fracNew1bynum']=[sum((ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x])&\
                         (ddates['firstLoan']==True))*1./
                         sum((ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x]))
                         for x in monthlyDat.index]
#amt of funding requested by returning borrowers
monthlyDat['amtReqRet1']=[sum(ddates['reqdamt'][(ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x])&\
                         (ddates['firstLoan']==False)]) for x in monthlyDat.index]
#amt of funding requested by new borrowers
monthlyDat['amtReqNew1']=[sum(ddates['reqdamt'][(ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x])&\
                         (ddates['firstLoan']==True)]) for x in monthlyDat.index]
#total number of loans requested
monthlyDat['numReq1']=[sum((ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x]))\
                          for x in monthlyDat.index]
#total number of loans requested by new borrowers
monthlyDat['numReqNew1']=[sum((ddates['applydate']<monthlyDat['per1End'][x])&
                         (ddates['applydate']>monthlyDat['per1Start'][x])&\
                         (ddates['firstLoan']==True)) for x in monthlyDat.index]
```

```python
#Calculate some specific metrics of interest to Julia
pm=3 # number of months to include in recent period
recentMon=monthlyDat[-pm:] #calculate metrics only for the most recent period P6M
print "Values for previous %i months:" %(pm) 
print "Request/Accept: %f" % (sum(recentMon['amtReqRet1'])/sum(recentMon['amtAccept1']))
print "Request/Bid: %f"%(sum(recentMon['amtReqRet1'])/sum(recentMon['amtBid1']))
print "Mean loan size: %f" %(sum(recentMon['amtReqNew1'])/sum(recentMon['numReqNew1']))
print "# to accept (lower): %i"%((mean(recentMon['amtAccept1'])*\
                                  (1+(.15*mean(recentMon['fundFractionExpire1'])))\
                                  -mean(recentMon['amtReqRet1']))\
        /(sum(recentMon['amtReqNew1'])/sum(recentMon['numReqNew1'])))
print "# to accept (upper): %i"%((mean(recentMon['amtBid1'])*\
                                  (1+(.15*mean(recentMon['fundFractionExpire1'])))\
                                  -mean(recentMon['amtReqRet1']))\
        /(sum(recentMon['amtReqNew1'])/sum(recentMon['numReqNew1'])))
```

```python
# Construct table containing total due, total paid, last payment due date for each loan,  
#   from 'repaymentschedule' table.
#   Also include conversion rate, and loan status ID
# Can you tell this happened after I spent more time learning mySQL? 

db = mdb.connect(user="root", host="localhost", db="zidisha", charset='utf8')
loanInfo = pd.read_sql("SELECT LA.loanid,LA.active,RS.totalsched,\
                    RS.totalpaid,RS.lastduedate,T.convrate\
                    FROM loanapplic LA \
                    JOIN (SELECT rtemp.loanid,SUM(rtemp.amount) \
                    AS totalsched,\
                    SUM(rtemp.paidamt) AS totalpaid,\
                    MAX(rtemp.duedate) AS lastduedate\
                    FROM repaymentschedule rtemp  \
                    GROUP BY loanid) AS RS\
                    ON LA.loanid=RS.loanid \
                    JOIN (SELECT t.loanid, AVG(t.conversionrate) as convrate\
                    FROM transactions t \
                    GROUP BY t.loanid) T\
                    ON T.loanid=LA.loanid",  db)
```

```python
# Since I had already calculated 'firstLoan' using pandas, didn't seem worth the SQL-ninjary to 
#    re-calculate so that I could combine everything into the query above. Just merged the info
#    below using pandas. 

loanRepayment=pd.merge(loanInfo,ddates[['loanid','firstLoan','reqdamt','borrowerid']],\
                       left_on='loanid',right_on='loanid',how='left')
# Dates in the Zidisha database sometimes come in different formats, so best to int them
loanRepayment['lastduedate']=[int(x) for x in \
                              loanRepayment['lastduedate']]
# 'totalpaid' and 'totalsched' are in local currancy, so converted to dollars
loanRepayment['totalpaidD']=loanRepayment['totalpaid']/\
                                loanRepayment['convrate']
loanRepayment['totalschedD']=loanRepayment['totalsched']/\
                                loanRepayment['convrate']   
```

```python
# How does the # of defaults vary with requested loan size

stdate=time.mktime(datetime.datetime(2013,7,1).timetuple())
enddate=time.mktime(datetime.datetime(2014,7,1).timetuple())

# This is a quick-and-dirty histogram construction. Zidisha wanted specific bins
histbins=[0.,51.,151.,501.,1001.,5001.]
loanRepayment['B1']=[x<histbins[1] for x in loanRepayment['reqdamt']]
loanRepayment['B2']=[(x>histbins[1]) &(x<histbins[2]) for x in \
                    loanRepayment['reqdamt']]
loanRepayment['B3']=[(x>histbins[2]) &(x<histbins[3]) for x in\
                    loanRepayment['reqdamt']]
loanRepayment['B4']=[(x>histbins[3]) &(x<histbins[4]) for x in\
                    loanRepayment['reqdamt']]
loanRepayment['B5']=[(x>histbins[4]) &(x<histbins[5]) for x in\
                    loanRepayment['reqdamt']]

Rdates=loanRepayment[(loanRepayment['firstLoan']==False) & \
              (loanRepayment['lastduedate']>stdate)&\
              (loanRepayment['lastduedate']<(enddate))]
fig,a=subplots(4)
h1=a[0].hist(list(Rdates['totalschedD'][(Rdates['active']==5)\
                                    | (Rdates['active']==3)]),histbins)
h2=a[1].hist(list(Rdates['totalschedD'][Rdates['active']==5]),histbins)

RdatesNew=loanRepayment[(loanRepayment['firstLoan']==True) & \
              (loanRepayment['lastduedate']>stdate)&\
              (loanRepayment['lastduedate']<enddate)]
histbins=[0,51,151,501,1001,5001]
h1N=a[2].hist(list(RdatesNew['totalschedD'][(RdatesNew['active']==5)\
                                    | (RdatesNew['active']==3)]),histbins)
h2N=a[3].hist(list(RdatesNew['totalschedD'][RdatesNew['active']==5]),histbins)
```
